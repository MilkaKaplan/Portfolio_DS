```python
# copydata_characteristic_attributes_mapping
def remove_punc(copydata_characteristic_attributes_mapping):
  pattern = r'[' + string.punctuation + ']'
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].map(lambda m:re.sub(pattern," ",m))
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].map(lambda m:re.sub(pattern," ",m))
  return copydata_characteristic_attributes_mapping


def lower(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].map(lambda m:m.lower())
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].map(lambda m:m.lower())
  return copydata_characteristic_attributes_mapping


def tokenization(copydata_characteristic_attributes_mapping):
    tokens = re.split(' ', copydata_characteristic_attributes_mapping)
    return tokens

def token(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']= copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: tokenization(x))
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']= copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: tokenization(x))
  return copydata_characteristic_attributes_mapping

#from nltk.corpus import stopwords
# russian_stopwords = stopwords.words("russian")

sw_russian=nltk.corpus.stopwords.words('russian')

def remove_SW_russian(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: [item for item in x if item not in sw_russian])
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: [item for item in x if item not in sw_russian])
  return copydata_characteristic_attributes_mapping

sw_english=nltk.corpus.stopwords.words('english')

def remove_SW_english(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: [item for item in x if item not in sw_english])
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: [item for item in x if item not in sw_english])
  return copydata_characteristic_attributes_mapping

def remove_digits(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: [item for item in x if not item.isdigit()])
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: [item for item in x if not item.isdigit()])
  
  return copydata_characteristic_attributes_mapping


lemmatizer = WordNetLemmatizer()

def lemmatize(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: [lemmatizer.lemmatize(item) for item in x])
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: [lemmatizer.lemmatize(item) for item in x])
 
  return copydata_characteristic_attributes_mapping


def remove_empty_tokens(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: [item for item in x if item !=''])
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: [item for item in x if item !=''])
  
  return copydata_characteristic_attributes_mapping


def remove_single_letters(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: [item for item in x if len(item) > 1])
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: [item for item in x if len(item) > 1])
  
  return copydata_characteristic_attributes_mapping


def detoken(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']= copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: TreebankWordDetokenizer().detokenize(x))
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']= copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: TreebankWordDetokenizer().detokenize(x))
  
  return copydata_characteristic_attributes_mapping

def replace_spaces(x,space,second):
  result = x.replace(space, second)
  return result
def remove_space(copydata_characteristic_attributes_mapping):
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x']= copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'].apply(lambda x: replace_spaces(x,'  ',' '))
  copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y']= copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'].apply(lambda x: replace_spaces(x,'  ',' '))
  
  return copydata_characteristic_attributes_mapping

def count_vcr_characteristic_attributes_mapping():
  for i in range(len(copydata_characteristic_attributes_mapping)):
    doc1_characteristic_attributes_mapping =copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'][i]
    doc2_characteristic_attributes_mapping =copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'][i]
    docs_characteristic_attributes_mapping =(doc1_characteristic_attributes_mapping ,doc2_characteristic_attributes_mapping)
    matrix_characteristic_attributes_mapping = CountVectorizer().fit_transform(docs_characteristic_attributes_mapping)
    cosine_sim_characteristic_attributes_mapping = cosine_similarity(matrix_characteristic_attributes_mapping[0], matrix_characteristic_attributes_mapping[1])
    similarity_characteristic_attributes_mapping.append(cosine_sim_characteristic_attributes_mapping)
  return similarity_characteristic_attributes_mapping

def similarity_fn_characteristic_attributes_mapping():
  for i in range(len(copydata_characteristic_attributes_mapping)):
    doc1_characteristic_attributes_mapping=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_x'][i]
    doc2_characteristic_attributes_mapping=copydata_characteristic_attributes_mapping['characteristic_attributes_mapping_y'][i]
    docs_characteristic_attributes_mapping=(doc1_characteristic_attributes_mapping,doc2_characteristic_attributes_mapping)
    tfidf_matrix_characteristic_attributes_mapping = tfidf_vectorizer.fit_transform(docs_characteristic_attributes_mapping)
    cosine_sim_characteristic_attributes_mapping = cosine_similarity(tfidf_matrix_characteristic_attributes_mapping[0], tfidf_matrix_characteristic_attributes_mapping[1])
    similarity_characteristic_attributes_mapping.append(cosine_sim_characteristic_attributes_mapping)
  return similarity_characteristic_attributes_mapping
```
