```python
#df = df[df['my_col'].notnull()].copy()
#df['my_col'] = df['my_col'].astype(int).astype(str)
def remove_punc(pair_characteristic):
    pattern = r'[' + string.punctuation + ']'
#    pair_characteristic = pair_characteristic[pair_characteristic[:,1].notnull()].copy()
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].map(lambda m:re.sub(pattern," ",m))
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].map(lambda m:re.sub(pattern," ",m))
    return pair_characteristic

def lower(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].map(lambda m:m.lower())
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].map(lambda m:m.lower())
    return pair_characteristic


def tokenization(pair_characteristic):
    tokens = re.split(' ', pair_characteristic)
    return tokens

def token(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: tokenization(x))
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: tokenization(x))
    return pair_characteristic

sw_russian=nltk.corpus.stopwords.words('russian')

def remove_SW_russian(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: [item for item in x if item not in sw_russian])
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: [item for item in x if item not in sw_russian])
    return pair_characteristic

sw_english=nltk.corpus.stopwords.words('english')

def remove_SW_english(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: [item for item in x if item not in sw_english])
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: [item for item in x if item not in sw_english])
    return pair_characteristic

#def remove_digits(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: [item for item in x if not item.isdigit()])
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: [item for item in x if not item.isdigit()])
    return pair_characteristic

lemmatizer = WordNetLemmatizer()

def lemmatize(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: [lemmatizer.lemmatize(item) for item in x])
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: [lemmatizer.lemmatize(item) for item in x])
    return pair_characteristic

def remove_empty_tokens(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: [item for item in x if item !=''])
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: [item for item in x if item !=''])
    return pair_characteristic


#def remove_single_letters(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: [item for item in x if len(item) > 1])
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: [item for item in x if len(item) > 1])
    return pair_characteristic


def detoken(pair_characteristic):
    pair_characteristic.iloc[:,1]= pair_characteristic.iloc[:,1].apply(lambda x: TreebankWordDetokenizer().detokenize(x))
    pair_characteristic.iloc[:,2]= pair_characteristic.iloc[:,2].apply(lambda x: TreebankWordDetokenizer().detokenize(x))
    return pair_characteristic

def replace_spaces(x,space,second):
    result = x.replace(space, second)
    return result

def remove_space(pair_characteristic):
    pair_characteristic.iloc[:,1]=pair_characteristic.iloc[:,1].apply(lambda x: replace_spaces(x,'  ',' '))
    pair_characteristic.iloc[:,2]=pair_characteristic.iloc[:,2].apply(lambda x: replace_spaces(x,'  ',' '))
    return pair_characteristic

#def count_vcr_characteristic(pair_characteristic):
#    for i in range(len(pair_characteristic)):
#        doc1_characteristic =pair_characteristic.iloc[:,1][i]
#        doc2_characteristic =pair_characteristic.iloc[:,2][i]
#        docs_characteristic =(doc1_characteristic,doc2_characteristic)
 #       matrix_characteristic = CountVectorizer().fit_transform(docs_characteristic)
 #       cosine_sim_characteristic = cosine_similarity(matrix_characteristic[0], matrix_characteristic[1])
 #       similarity_characteristic.append(cosine_sim_characteristic)
 #       return similarity_characteristic

#def similarity_fn_characteristic():
#    for i in range(len(pair_characteristic)):
 #       doc1_characteristic=pair_characteristic.iloc[:,1][i]
 #       doc2_characteristic=pair_characteristic.iloc[:,2][i]
 #       docs_characteristic=(doc1_characteristic,doc2_characteristic)
 #       tfidf_matrix_characteristic = tfidf_vectorizer.fit_transform(docs_characteristic)
 #       cosine_sim_characteristic = cosine_similarity(tfidf_matrix_characteristic[0], tfidf_matrix_characteristic[1])
 #       similarity_characteristic.append(cosine_sim_characteristic)
 #       return similarity_characteristic
```

```python
#функция для нахождения косинусной близости эмбедингов изображений
def similarity(embeddings):
    cos_sim = []
    for i in range(len(embeddings)):
        embedding1 = embeddings.iloc[:,1][i][0].reshape(1, -1)
        embedding2 = embeddings.iloc[:,2][i][0].reshape(1, -1)
        similarity = cosine_similarity(embedding1, embedding2)
        cos_sim.append(similarity[0][0])
    return cos_sim
```

```python
#функция для нахождения косинусной близости эмбедингов названий
def similarity_bert(embeddings):
    cos_sim = []
    for i in range(len(embeddings)):
        embedding1 = embeddings.iloc[:,1][i].reshape(1, -1)
        embedding2 = embeddings.iloc[:,2][i].reshape(1, -1)
        similarity = cosine_similarity(embedding1, embedding2)[0][0]
        cos_sim.append(similarity)
    return cos_sim
```
